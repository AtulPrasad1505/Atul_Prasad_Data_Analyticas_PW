Q17.  Explain the difference between apply() and map() in Pandas?

Q21.  Why is NumPy’s array slicing faster than Python’s list slicing?
  - NumPy’s array slicing is significantly faster than Python’s built-in list slicing because of efficient memory management, optimized C-based operations, and the way NumPy handles data in contiguous memory blocks. NumPy arrays are stored in contiguous memory blocks, unlike Python lists, which store references to objects at different memory locations. Slicing in NumPy returns a "view" (not a copy), meaning no new memory is allocated.NumPy avoids unnecessary memory allocation and speeds up slicing. NumPy’s slicing and indexing use optimized C and Fortran-level operations, whereas Python lists require loops and additional memory allocations.

```python
#Q4.  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10),
    'D': np.random.rand(10)
}

df = pd.DataFrame(data)
corr_matrix = df.corr()
print(corr_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Correlation Matrix Heatmap")
plt.show()


```

```python
#Q13.  Create a 3D scatter plot using Plotly.


import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)
df = pd.DataFrame({
    'X': np.random.randn(100),
    'Y': np.random.randn(100),
    'Z': np.random.randn(100),
    'Category': np.random.choice(['A', 'B', 'C'], 100)
})

fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Category',
                    title='3D Scatter Plot with Category')

fig.show()
```
