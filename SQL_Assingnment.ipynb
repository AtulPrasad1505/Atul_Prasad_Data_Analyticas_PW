2. Explain the purpose of constraints and how they help maintain data integrity in a database. Provide examples of common types of constraints.
  - Purpose of Constraints in Databases: Constraints are rules applied to database columns to enforce data integrity, accuracy, and consistency. They help ensure that the data entered into a database follows predefined rules and business logic, preventing invalid or inconsistent data from being stored. By doing so, constraints maintain the reliability and quality of data across the database system.

  - How Constraints Help Maintain Data Integrity:
  1. Prevent Invalid Data Entry: Constraints ensure only valid data is inserted into a table.
  2. Enforce Relationships: They maintain logical connections between tables (e.g., foreign keys).
  3. Ensure Uniqueness and Accuracy: By enforcing unique and required values where necessary.
  4. Support Business Rules: They help implement real-world rules (e.g., prices can't be negative).

  - Eg : Not Null, UNIQUE, Primary Key, Forigen Key, Check, Default.

3.Why would you apply the NOT NULL constraint to a column? Can a primary key contain NULL values? Justify your answer.
  - The NOT NULL constraint is used to ensure that a column always contains a  value, i.e., it cannot be left empty (NULL). This is important when the column holds essential information that the database or application logic depends on.
  - A primary key cannot contain NULL values.  A primary key must uniquely identify each row in a table. NULL represents an unknown or missing value, and allowing NULL would violate the principle of uniqueness. Primary keys are often used in joins and references (foreign keys), and NULLs would break referential integrity.

4. Primary keys are often used in joins and references (foreign keys), and NULLs would break referential integrity.
  - Primary keys are used to uniquely identify each record in a table, and foreign keys refer to those primary keys to create relationships between tables. If a primary key were allowed to be NULL, it would introduce uncertainty—because NULL means "unknown," it cannot reliably serve as a unique identifier.
  In terms of referential integrity, this causes problems because:
  A foreign key referencing a NULL primary key would have no clear match.
  You couldn't enforce that all foreign keys actually point to valid, existing records.

3. Third Normal Form (3NF):Identify a table in Sakila that violates 3NF. Describe the transitive dependencies present and outline the steps to normalize the table to 3NF.
  -  A relation (table) is in Third Normal Form (3NF) if:

	It is already in Second Normal Form (2NF), and

	There are no transitive dependencies, i.e.,

	No non-prime attribute depends on another non-prime attribute.
	In other words, all non-key attributes must depend only on the primary key, and not on other non-key attributes.

	- Let us consider a modified version of the customer table from the Sakila database. Suppose it is structured as follows:

	Column	Description
	customer_id	Primary Key
	address_id	Foreign Key referencing address
	city	Name of the city (derived from address)
	country	Name of the country (derived from city)

	- In this design:

	customer_id is the primary key.

	address_id determines the city.

	The city determines the country.

	- Transitive Dependencies Present
	Here’s how the transitive dependencies appear:

	customer_id → address_id

	address_id → city

	city → country

	- Therefore, customer_id → city → country is a transitive dependency, because country depends on city, which itself depends on address_id, which depends on customer_id.

	- This violates 3NF because a non-key attribute (country) is not directly dependent on the primary key, but instead depends on another non-key attribute (city).

	- Steps to Normalize the Table to 3NF
	To eliminate transitive dependencies and bring the table into 3NF, follow these steps:

	1. Create a country table
	This table stores unique country names:
	country(country_id, country_name)
	2. Create a city table
	Each city belongs to a country:
	city(city_id, city_name, country_id)
	3. Update the address table
	Store only the reference to the city:
	address(address_id, address_details, city_id)
	4. Keep the customer table clean
	Now, the customer table includes only attributes that directly relate to the customer:
	customer(customer_id, address_id, first_name, last_name, email)
	The design now eliminates all transitive dependencies.

	Every non-key attribute depends only on the primary key of its table.

	The tables are now fully normalized to Third Normal Form (3NF).

unnormalized form up to at least 2NF.
  - Let’s assume the payment table is initially in Unnormalized Form (UNF).

  - Unnormalized Form (UNF)
  In UNF:The table may contain repeating groups or non-atomic attributes (multiple values in one field).
Data may have redundancy and lack a primary key.

  - Hypothetical UNF Structure: payment(payment_id, customer_id, customer_name, payment_date, amount, staff_id, staff_name)

  All attribute values are atomic (indivisible).

  There are no repeating groups or multivalued fields.

  - Transformation to 1NF:Ensure each column holds only atomic values.Remove repeating groups. Create separate tables for customer and staff entities.

  - Result:
  payment retains only relevant payment data and foreign keys. customer and staff data are stored in separate tables.
  All fields are atomic and logically separated.

  - Second Normal Form (2NF)
 A table is in 2NF if: It is already in 1NF. It has no partial dependency — that is, every non-key attribute is fully functionally dependent on the entire primary key.

  - Example of Partial Dependency: If a table has a composite primary key (e.g., customer_id, payment_date), and attributes like customer_name depend only on customer_id, not on the full key, then: customer_name is partially dependent on the primary key.

  - This violates 2NF.
  - Transformation to 2NF:Move partially dependent attributes (like customer_name) into a separate table.
  Create relationships between the new tables using foreign keys.

  - Result: The payment table now includes only data that depends fully on the primary key. Customer and staff details are stored in separate, normalized tables.
