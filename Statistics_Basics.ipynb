Q1.  Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales?
  - 1. Qualitative Data is also known as Categorical Data - This type of data represents characteristics, descriptions, or attributes that cannot be measured numerically. Instead, it classifies data into categories. Eg - Customer feedback (Excellent, Good, Poor).
  There are Two Tyoes of Qualitative Data:
  (1) Nominal Scale: Data with categories that have no inherent order.
  (2) Ordinal Scale: Data with a meaningful order but no consistent difference between values.

  2. Quantitative Data is also known as Numerical Data - This type of data represents measurable quantities and can be counted or expressed in numerical values. Eg - Revenue generated from events, Number of glasses in a champagne tower.
  There are Two Tyoes of Qualitative Data:
  (1) Interval Scale: Numerical data where the difference between values is meaningful, but there is no true zero.
  (2) Ratio Scale: Numerical data with a true zero, meaning ratios can be compared.

Q3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
- Dispersion refers to the extent to which data points in a dataset vary or spread out from the central value (such as the mean or median). It helps in understanding how much the values deviate from each other and provides insight into the reliability and consistency of the data.
  1. Variance : Variance measures the average squared deviation from the mean.
  A higher variance indicates greater spread in the data, while a lower variance suggests the values are closer to the mean. Variance provides a measure of dispersion but is in squared units, making it less intuitive for direct interpretation.
  2. Standard Deviation : Standard deviation is the square root of variance.
  Unlike variance, which is in squared units, the standard deviation is in the same units as the original data, making it more interpretable.
  Standard deviation brings the measure back to the original units, making it more useful for comparison.

Q6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
- Skewness is a measure of the asymmetry of a probability distribution. It describes how data points are distributed around the mean—whether they are more concentrated on one side or spread out unevenly. A perfectly symmetric distribution like a normal distribution has zero skewness.
> Types of Skewness : 1. Positive Skew (Right-Skewed), 2. Negative Skew (Left-Skewed), 3. Zero Skew (Symmetrical Distribution)

  How Skewness Affects Data Interpretation : (1) Affects Measures of Central Tendency: In skewed distributions, the mean is pulled in the direction of the skew, making the median a better measure of central tendency.
  (2) Influences Decision Making: Understanding skewness helps in making better predictions. For example, in finance, a right-skewed return distribution suggests a potential for large gains, whereas a left-skewed one indicates a risk of significant losses.
  (3) Impacts Statistical Tests: Many parametric tests assume normality; highly skewed data may require transformations or non-parametric methods.


Q7. What is the interquartile range (IQR), and how is it used to detect outliers?
- The Interquartile Range (IQR) is a measure of statistical dispersion, representing the middle 50% of a dataset. It is calculated as:
> IQR=Q3 − Q1
  
  where:

  Q₁ (First Quartile) = The 25th percentile (median of the lower half of the data).

  Q₃ (Third Quartile) = The 75th percentile (median of the upper half of the data).
  
  IQR helps understand the spread of data by showing how concentrated the middle portion is.

  How is IQR Used to Detect Outliers:
  Outliers are extreme values that lie far from the majority of the data points. The IQR method identifies outliers using the following rule:

  1. Calculate the lower and upper boundaries:

    Lower Bound = 𝑄1−1.5×𝐼𝑄𝑅Q 1​ −1.5×IQR

    Upper Bound = 𝑄3+1.5×𝐼𝑄𝑅Q 3​ +1.5×IQR
  2. Identify Outliers:

  Any data point below the lower bound or above the upper bound is considered an outlier.


Q8. Discuss the conditions under which the binomial distribution is used.
- The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials, where each trial has two possible outcomes: success or failure. It is commonly used in scenarios like coin flips, quality control, and medical trials.

  To use the binomial distribution, the following four conditions must be met:
  1. Fixed Number of Trials (n)
  2. Only Two Possible Outcomes (Success or Failure)
  3. Constant Probability of Success (p).
  4. Independent Trials.

  When to Use the Binomial Distribution:
  (1) Tossing a Coin: Counting the number of heads in 10 coin flips.
  (2) Defective Products in Manufacturing: Finding the probability of 3 defective items in a batch of 20 when the defect rate is 5%.
  (3) Medical Trials: Estimating how many patients out of 50 will respond to a new drug.


Q9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
- The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric around the mean. It is widely used in statistics because many natural phenomena (e.g., height, IQ scores, test scores) follow this pattern.
  > Key Properties:
  1. Symmetry : The normal distribution is perfectly symmetric about its mean 𝜇
  2. Mean, Median, and Mode are Equal
  3. Bell-Shaped Curve
  4. Defined by Mean (μ) and Standard Deviation (σ)
  5. Total Area Under the Curve = 1

  > The Empirical Rule (68-95-99.7 Rule):
  The Empirical Rule provides a quick way to estimate how data is distributed in a normal distribution.
  >>  68% of data falls within 1 standard deviation (𝜇±1𝜎μ±1σ).

  >>  95% of data falls within 2 standard deviations (𝜇±2𝜎μ±2σ).

  >> 99.7% of data falls within 3 standard deviations (𝜇±3𝜎μ±3σ).

Q11.Explain what a random variable is and differentiate between discrete and continuous random variables.
- A random variable is a numerical outcome of a random experiment. It assigns a number to each possible event in a sample space. Random variables are essential in probability and statistics as they help quantify uncertainty.
> Types of Random Variables :
  1. Discrete Random Variables:
  A discrete random variable takes on a countable number of distinct values.
  Eg : (1) The number of heads in 5 coin flips (values: 0, 1, 2, 3, 4, 5).
  (2) The number of customers arriving at a store in an hour.

  2. Continuous Random Variables :
  A continuous random variable can take on an infinite number of values within a given range. It is typically the result of measuring something.
  > Eg: The height of students in a class. and The time it takes to complete a task.
